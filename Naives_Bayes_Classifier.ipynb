```python
import pandas as pd
df=pd.read_csv("iris.csv")
df
# use case : classifying different species of iris flower dataset based on sepal and petal length and width
# using Naives Bayes Theorem
# setosa(small),verginica(medium),versicolor(large)
```

```python
# Feature Extraction
x=df.loc[:,["sepal.length","sepal.width","petal.length","petal.width"]]
x
```

```python
# Feature Extraction
y=df["variety"]
df
```

```python
df.describe()
```

```python
#step-5 : train test split model 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#step-6 : EDA2 
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
```

```python
from sklearn.naive_bayes import GaussianNB
model1=GaussianNB()
model1.fit(x_test,y_test)
```

```python
#step-7 : model building and model evaluation
y_pred=model.predict(x_train)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_train,y_pred)
cm
```

```python
print(accuracy_score(y_train,y_pred))
```

```python
import seaborn as sns
sns.heatmap(cm,annot=True,xticklabels=['Setosa', 'Verginica', 'Versicolor'],
           yticklabels=['Setosa', 'Virginica','Versicolor']) 
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_train,y_pred))
```

```python
y_pred1=model1.predict(x_test)
y_pred1
```

```python
print(accuracy_score(y_test, y_pred1))
```

```python
cm1=confusion_matrix(y_test,y_pred1)
cm1
```

```python
sns.heatmap(cm1,annot=True,xticklabels=['Setosa', 'Verginica', 'Versicolor'],
           yticklabels=['Setosa', 'Virginica','Versicolor']) 
```

```python
print(classification_report(y_test,y_pred1))
```
